### Introdução

A US11 visa recolher dados do portal do utilizador sobre a utilização do parque, com o intuito de compreender o uso do parque pelas diferentes faixas etárias. Esta user story recebe um arquivo .csv como entrada, contendo três variáveis: a faixa etária dos utilizadores do parque, se estes recomendam o parque e quantas vezes utilizam o parque por mês. O output é a proporção de utilizadores que recomendam o parque em cada faixa etária e um boxplot com a frequência de visitas por faixa etária.

#### Proporção de utilizadores que recomendam o parque por faixa etária

Input: arquivo .csv, faixa etária, recomendações

Output: valor numérico

Com base nestes dados, o programa realiza a contagem de recomendações por faixa etária. Após agrupar as idades com as recomendações (através do groupby Escolha - Y/N), o programa calcula a proporção de utilizadores que recomendam o parque em cada faixa etária.

#### Frequência de visitas por faixa etária

Input: arquivo .csv, faixa etária, número de visitas por mês

Output: Boxplot

Com estes dados, o programa gera um boxplot, que é um diagrama de extremos e quartis (utilizando a biblioteca matplotlib.pyplot), representando a frequência de visitas mensais por faixa etária. O boxplot inclui a representação gráfica dos valores mínimos, máximos e medianos de visitas mensais por faixa etária.

### Código e Resultados

```python
#US11
from scipy import stats
import csv
import matplotlib.pyplot as plt
import pandas as pd

# Guarda o ficheiro csv num DataFrame (df)
CAMINHO_FICHEIRO = 'Inquiry.csv'
df = pd.read_csv(CAMINHO_FICHEIRO, sep=';', header=None, names=['Escalao', 'Y/N', 'Visits'])

# Converter a coluna 'Visits' para o tipo numérico
df['Visits'] = pd.to_numeric(df['Visits'], errors='coerce')

# Tipos de Variáveis
print("Tipo de cada uma das variáveis:")

print("\nFaixa etária?-")
print("\nVariável Qualitativa Ordinal")
print("\nRecomendaria a utilização deste parque a outras pessoas? -")
print("\nVariável Qualitativa Nominal") 
print("\nQuantas vezes por mês visitou o parque? -")
print("\nVariável Quantitativa Discreta") 

# Contagem de recomendações por faixa etária
# Agrupar as idades com as recomendações (Y)
# Calcular a proporção de usuários que recomendam o parque (por faixa etária)
proporcaoPorIdades = df.groupby('Escalao')['Y/N'].apply(lambda x: round((x == 'Y').mean(), 2)) # Resultado arredondado a duas casas décimais

print("\nO índice de aprovação entre as crianças para o parque é:", proporcaoPorIdades['1'])
print("O índice de aprovação entre os adultos para o parque é:", proporcaoPorIdades['2'])
print("O índice de aprovação entre os séniores para o parque é:", proporcaoPorIdades['3'])

# Boxplot com a frequência de visitas por faixa etária  
df.boxplot(column='Visits', by='Escalao', grid=True, color=dict(boxes='blue', whiskers='green', medians='red', caps='gray'), vert=True, showfliers=True)
plt.title('Frequência de Visitas ao Parque por Faixa Etária')
plt.xlabel('Faixa Etária')
plt.ylabel('Frequência de Visitas')
plt.xticks([1, 2, 3], ['Crianças', 'Adultos', 'Séniores'])
plt.tight_layout()  # Adiciona espaçamento
plt.show()
```
