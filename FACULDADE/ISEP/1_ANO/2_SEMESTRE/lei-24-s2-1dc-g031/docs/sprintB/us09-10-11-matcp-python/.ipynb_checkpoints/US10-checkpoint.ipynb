### US10

## Introdução

A US10 tem como objetivo identificar os equipamentos do parque mais frequentemente utilizados diariamente, a fim de compreender as preferências das pessoas que frequentam o parque. Esta user story recebe como entrada um arquivo contendo as escolhas das pessoas e gera como saída um gráfico de pizza que mostra as percentagens de utilização de cada equipamento.

### Percentagem de Utilização dos Equipamentos do Parque
Input: arquivo .csv, nome do equipamento

Output: gráfico de pizza

Com base nos dados fornecidos, o programa gera como saída um gráfico de pizza (utilizando a biblioteca matplotlib.pyplot) que exibe as percentagens de utilização dos diversos equipamentos do parque.

## Código e Resultados

```python
#US10
import matplotlib.pyplot as plt
import pandas as pd

# Definir o caminho do arquivo como constante
CAMINHO_FICHEIRO = 'EquipmentUsed.csv'

# Ler o arquivo CSV em um DataFrame
df = pd.read_csv(CAMINHO_FICHEIRO, sep=';')

# Contar o número total de equipamentos
quantidade_de_equipamentos = len(df)

# Contar o número de vezes que cada equipamento foi usado
total_de_equipamentos = df["Equipment"].value_counts()

# Calcular as percentagens de utilização dos equipamentos
percentagem_de_utilizacao = (total_de_equipamentos / quantidade_de_equipamentos) * 100

# Definir cores manualmente
cores = ['blue', 'lightblue', 'green', 'lightgreen', 'purple', 'pink', 'grey']

# Plotar o gráfico de pizza
plt.pie(total_de_equipamentos, labels=total_de_equipamentos.index, autopct="%1.2f%%", colors=cores)
plt.title("Gráfico da Utilização dos Equipamentos do Parque")
plt.axis('equal')  # Para ser um gráfico redondo
plt.subplots_adjust(top=1.3)  # Garante espaço entre o título e o gráfico
plt.show()  # Mostrar o gráfico
```
